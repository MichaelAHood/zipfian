```python
import pandas as pd
```

```python
Salesperson = {"ID" : pd.Series([1, 2, 5, 7, 8, 11]),
               "Name": pd.Series(["Abe", "Bob", "Chris", "Dan", "Ken", "Joe"]),
               "Age" : pd.Series([61, 34, 34, 41, 57, 38]),
               "Salary" : pd.Series([140000, 44000, 40000, 52000, 115000, 38000])
               }
```

```python
Salesperson = pd.DataFrame(Salesperson)
```

```python
Salesperson
```

```python
Orders = {"Number" : pd.Series([10, 20, 30, 40, 50, 60, 70]),
          "order_date" : pd.Series(["8/2/96", "1/30/99", "7/14/95", "1/29/98", "2/3/98", "3/2/98", "5/6/98"]),
          "cust_id" : pd.Series([4, 4, 9, 7, 6, 6, 9]),
          "salesperson_id" : pd.Series([2, 8, 1, 2, 7, 7, 7]),
          "Amount" : pd.Series([540, 1800, 460, 2400, 600, 720, 150])
          }
```

```python
Orders = pd.DataFrame(Orders)
```

```python
Orders
```

```python
# Now that we have the data in dataframes we can join them.

# Since Salesperson.ID and Orders.salesperson_id are the same thing
# we rename both to "salesperson_id" and create a merge on column.

Salesperson = Salesperson.rename(columns={"ID" : "salesperson_id"})
```

```python
joined_df = pd.merge(Orders, Salesperson, on="salesperson_id")
```

```python
joined_df
```

```python
# Group by Name and set the 
# size of each, in this case the number of sales,
# to a new dataframe called grouped

grouped = joined_df.groupby(joined_df.Name).size()
```

```python
# Return the Name where there are more than 1 occurence

grouped[grouped > 1]
```
